```python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

# Load GloVe vectors
glove_vectors_path = 'glove.6B.100d.txt'

def load_glove_vectors(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        word_to_vec = {}
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], dtype='float32')
            word_to_vec[word] = vector
    return word_to_vec

glove_vectors = load_glove_vectors(glove_vectors_path)

# Load the dataset
df = new_df

# Preprocess the tags
def preprocess_tags(tag):
    stop_words = set(stopwords.words('english'))
    tag = tag.lower()  # Convert to lowercase
    tag = tag.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    word_tokens = word_tokenize(tag)
    filtered_tag = [word for word in word_tokens if word not in stop_words]
    return ' '.join(filtered_tag)

df['Tag'] = df['Tag'].apply(preprocess_tags)

# Convert tags to vectors using GloVe vectors
def tag_to_vector(tag):
    tokens = tag.split()
    vectors = [glove_vectors[word] for word in tokens if word in glove_vectors]
    if vectors:
        tag_vector = np.mean(vectors, axis=0)
        # Normalize the vector
        norm = np.linalg.norm(tag_vector)
        if norm != 0:
            tag_vector /= norm
        return tag_vector
    else:
        return np.zeros(len(list(glove_vectors.values())[0]))

df['Tag_Vector'] = df['Tag'].apply(tag_to_vector)

# Function to recommend similar games
def recommend_similar_games(user_game, top_n=5):
    filtered_df = df[df['Name'] == user_game]
    if filtered_df.empty:
        print(f"Game '{user_game}' not found in the dataset.")
        return []

    user_game_vector = filtered_df['Tag_Vector'].values[0]
    similarity_scores = []

    for _, row in df.iterrows():
        tag_vector = row['Tag_Vector']
        similarity = euclidean_distances([user_game_vector], [tag_vector])[0][0]
        similarity_scores.append(similarity)

    df['Similarity'] = similarity_scores

    # Filter out the user's game from the recommended list
    recommended_games = df[df['Name'] != user_game].sort_values(by='Similarity', ascending=True)['Name'].head(top_n)
    return recommended_games.tolist()


# Example usage
user_input_game = 'Galactic Bowling'
recommended_games = recommend_similar_games(user_input_game)
if recommended_games:
    print("Recommended Games:")
    for idx, game in enumerate(recommended_games):
        print(f"{idx+1}. {game}")
```
