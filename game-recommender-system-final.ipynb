```python
def remove_irregularities(tag):
    # Remove special characters
    tag = ''.join(e for e in tag if e.isalnum() or e.isspace())
    # Remove extra whitespace
    tag = ' '.join(tag.split())
    return tag

# Apply the cleaning function to each tag in the 'Tag' column of the DataFrame 'new_df'
new_df['Tag'] = new_df['Tag'].apply(remove_irregularities)
```

```python
#GLOVE
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Load GloVe vectors
glove_vectors_path = 'glove.6B.300d.txt'

def load_glove_vectors(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        word_to_vec = {}
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], dtype='float32')
            word_to_vec[word] = vector
    return word_to_vec

glove_vectors = load_glove_vectors(glove_vectors_path)

# Load the dataset
df = new_df

# Preprocess the tags
def preprocess_tags(tag):
    return ' '.join(tag.split())  # Example preprocessing, you can customize this as needed

df['Tag'] = df['Tag'].apply(preprocess_tags)

# Convert tags to vectors using GloVe vectors
def tag_to_vector(tag):
    tokens = tag.split()
    vectors = [glove_vectors[word] for word in tokens if word in glove_vectors]
    if vectors:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(len(list(glove_vectors.values())[0]))

df['Tag_Vector'] = df['Tag'].apply(tag_to_vector)

# Function to recommend similar games
def recommend_similar_games(user_game, top_n=5):
    filtered_df = df[df['Name'] == user_game]
    if filtered_df.empty:
        print(f"Game '{user_game}' not found in the dataset.")
        return []

    user_game_vector = filtered_df['Tag_Vector'].values[0]
    user_game_vector = np.squeeze(user_game_vector)  # Ensure it's not a multi-dimensional array
    similarity_scores = []

    for _, row in df.iterrows():
        tag_vector = np.squeeze(row['Tag_Vector'])
        similarity = cosine_similarity(user_game_vector.reshape(1, -1), tag_vector.reshape(1, -1))[0][0]
        similarity_scores.append(similarity)

    df['Similarity'] = similarity_scores

    # Filter out the user's game from the recommended list
    recommended_games = df[df['Name'] != user_game].sort_values(by='Similarity', ascending=False)['Name'].head(top_n)
    return recommended_games.tolist()


# Example usage
user_input_game = 'Forza Horizon 4'
recommended_games = recommend_similar_games(user_input_game)
if recommended_games:
    print("Recommended Games:")
    for idx, game in enumerate(recommended_games):
        print(f"{idx+1}. {game}")
```
